```python
df=df[df["parentesco_jefe_hogar"]=="jefe_hogar"]
```

```python
df=df[["edad","sexo","ingreso_mensual_aprox","viaje_noche_fuera","estado_civil","motivo_viaje","estrato"]]
```

```python
df.isna().sum()
```

```python
df.drop(columns=["motivo_viaje"],inplace=True)
```

```python
df["estrato"].fillna(df["estrato"].median(),inplace=True)
```

```python
df["ingreso_mensual_aprox"].fillna(df["ingreso_mensual_aprox"].median(),inplace=True)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates().reset_index(drop=True)
```

```python
df.replace({"viaje_noche_fuera":{"si":1,"no":0}},inplace=True)
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
y=df.pop("viaje_noche_fuera")
x=sm.add_constant(df.copy())
model=sm.OLS(y,x)
reg=model.fit()
reg.summary()
```
